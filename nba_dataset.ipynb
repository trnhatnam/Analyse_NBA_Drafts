```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from wordcloud import WordCloud
from sklearn.metrics import r2_score
```

```python
df_award = pd.read_csv('nba/Player Award Shares.csv')
df_award = df_award.rename(columns={'player' : 'player_name'})
```

```python
df_award
```

```python
df_career = pd.read_csv('nba/Player Career Info.csv')
df_career = df_career.rename(columns={'player' : 'player_name'})
df_career
```

```python
df_playoff = pd.read_csv('nba/Team Stats Per 100 Poss.csv')
df_playoff = df_playoff.rename(columns={'abbreviation' : 'team_abbreviation'})
df_playoff
```

```python
df_rookie = pd.read_csv('nba/nba_all_season_rookie_season.csv')
df_rookie
```

```python
# Fusionner les DataFrames en utilisant une fusion interne sur la colonne 'player_name'
df_rookie = pd.merge(df_rookie, df_career[['player_name','num_seasons' ,'hof']], on='player_name', how='inner')

# Renommer la colonne 'hof' pour 'hall of fame' si nécessaire
df_rookie = df_rookie.rename(columns={'hof': 'hall_of_fame'})
df_rookie
```

```python
df_rookie.loc[df_rookie['player_name'] == 'Rudy Gobert']
```

```python
# Grouper les données dans df_award par joueur pour vérifier si au moins une occurrence de "True" existe
player_award_status = df_award.groupby('player_name')['winner'].any().reset_index()

# Fusionner df_rookie avec les résultats du groupage
df_rookie = pd.merge(df_rookie, player_award_status, on='player_name', how='left')

# Remplacer les valeurs NaN par False dans la colonne 'winner'
df_rookie['winner'].fillna(False, inplace=True)

# Renommer la colonne 'winner' en 'indiv_award' si nécessaire
df_rookie = df_rookie.rename(columns={'winner': 'indiv_award'})
df_rookie
```

```python
df_rookie.loc[df_rookie['player_name'] == 'Rudy Gobert']
```

```python
# Convertir la colonne 'season' de df_rookie en int64 en extrayant les quatre premiers caractères
df_rookie['season_year'] = df_rookie['season'].str[:4].astype(int)
df_r_stock = df_rookie['season']
# Fusionner df_rookie avec df_playoff sur la colonne 'team_abbreviation' et 'season_year'
df_rookie = pd.merge(df_rookie, df_playoff[['team_abbreviation', 'season', 'playoffs']], 
                     left_on=['team_abbreviation', 'season_year'], 
                     right_on=['team_abbreviation', 'season'], 
                     how='left')

# Remplacer les valeurs NaN par False dans la colonne 'playoffs'
df_rookie['playoffs'].fillna(False, inplace=True)
df_rookie
```

```python
df_rookie.loc[df_rookie['player_name'] == 'Giannis Antetokounmpo']
```

```python
df_rookie.drop(columns=['season_y'], inplace=True)
df_rookie['season'] = df_r_stock
df_rookie
```
